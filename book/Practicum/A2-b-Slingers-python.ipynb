`````python
import numpy as np
import matplotlib.pyplot as plt

l = np.array([1, 2, 3, 4, 5]) #vul hier je waardes van l in gescheiden door een komma
T = np.array([1, 2, 3, 4, 5]) #vul hier je waardes van T in gescheiden door een komma

plt.plot(l, T, 'o')
plt.xlabel('Lengte slinger (m)')
plt.ylabel('Periode (s)')
plt.show()

T2 = T**2

plt.plot(l, T2, 'o')
plt.xlabel('Lengte slinger (m)')
plt.ylabel('Periode^2 (s^2)')
plt.show()
`````
